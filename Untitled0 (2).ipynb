```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sentences = [

"The batsman hit a six in the cricket match.",
"The bowler bowled a fast delivery.",
"The cricket team celebrated their victory.",
"The captain changed the field setting.",

"The chef cooked delicious pasta.",
"She baked a chocolate cake.",
"He chopped vegetables for soup.",

"Strong passwords protect accounts.",
"Cybersecurity experts monitor networks.",
"Encryption protects sensitive data."

]
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')

embeddings = model.encode(sentences)

print(embeddings.shape)
```

```python
similarity_matrix = cosine_similarity(embeddings)

print(similarity_matrix)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(similarity_matrix, annot=True, cmap="coolwarm")
plt.title("Cosine Similarity Heatmap")
plt.show()
```

```python
query = "The bowler took three wickets in one over"

query_embedding = model.encode([query])

similarities = cosine_similarity(query_embedding, embeddings)[0]

top_indices = similarities.argsort()[-2:][::-1]

print("Query:", query)

print("\nTop 2 similar sentences:")

for idx in top_indices:
    print(sentences[idx])
    print("Similarity:", similarities[idx])
```
